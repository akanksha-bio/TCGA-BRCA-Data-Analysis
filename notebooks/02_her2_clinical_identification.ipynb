# Task 2: HER2+ Clinical Identification

**Goal:** Identify HER2-positive patients from clinical data (IHC, FISH).

Steps:
1. Load and explore clinical data
2. Inspect HER2-related columns (IHC, FISH, etc.)
3. Define HER2+ status
4. Summarize cohort demographics
5. Save cleaned clinical data

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ..src.data_utils import load_clinical, save_processed
```

## 1. Load and explore clinical data

```python
clinical = load_clinical()
print(clinical.shape)
clinical.head()
```

```python
# Identify HER2-related columns
her2_cols = [c for c in clinical.columns if 'her2' in c.lower() or 'erbb' in c.lower() or 'fish' in c.lower()]
print('HER2-related columns:', her2_cols)
clinical[her2_cols].head(20)
```

## 2. Inspect IHC-HER2 column

```python
# TODO: update column name if different in actual data
ihc_col = 'IHC-HER2'  # adjust as needed
print(clinical[ihc_col].value_counts(dropna=False))
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
clinical[ihc_col].value_counts(dropna=False).plot(kind='bar', ax=ax, color='steelblue', edgecolor='white')
ax.set_title('IHC-HER2 Status Distribution')
ax.set_xlabel('IHC Status')
ax.set_ylabel('Count')
plt.tight_layout()
plt.savefig('../figures/02_ihc_distribution.png', dpi=150)
plt.show()
```

## 3. Define HER2+ status

```python
# Primary: IHC Positive = HER2+
# Optionally incorporate FISH if available
clinical['her2_positive'] = clinical[ihc_col].str.strip().str.lower() == 'positive'

print('HER2+ :', clinical['her2_positive'].sum())
print('HER2- :', (~clinical['her2_positive']).sum())
print('Missing:', clinical['her2_positive'].isna().sum())
```
